```python
# Useful Constants

# Those are separate normalised input features for the neural network
INPUT_SIGNAL_TYPES = [
    "body_acc_x_",
    "body_acc_y_",
    "body_acc_z_",
    "body_gyro_x_",
    "body_gyro_y_",
    "body_gyro_z_",
    "total_acc_x_",
    "total_acc_y_",
    "total_acc_z_"
]

# Output classes to learn how to classify
LABELS = [
    "WALKING", 
    "WALKING_UPSTAIRS", 
    "WALKING_DOWNSTAIRS", 
    "SITTING", 
    "STANDING", 
    "LAYING"
]

DATA_PATH = "./"

DATASET_PATH = DATA_PATH + "UCI HAR Dataset/"
print("\n" + "Dataset is now located at: " + DATASET_PATH)

TRAIN = "train/"
TEST = "test/"


# Load "X" (the neural network's training and testing inputs)

def load_X(X_signals_paths):
    X_signals = []
    
    for signal_type_path in X_signals_paths:
        file = open(signal_type_path, 'r')
        # Read dataset from disk, dealing with text files' syntax
        X_signals.append(
            [np.array(serie, dtype=np.float32) for serie in [
                row.replace('  ', ' ').strip().split(' ') for row in file
            ]]
        )
        file.close()
    
    return np.transpose(np.array(X_signals), (1, 2, 0))

X_train_signals_paths = [
    os.path.join(*[DATASET_PATH, TRAIN, "Inertial Signals/", signal+"train.txt"]) for signal in INPUT_SIGNAL_TYPES
]
X_test_signals_paths = [
    os.path.join(*[DATASET_PATH, TEST, "Inertial Signals/", signal+"test.txt"]) for signal in INPUT_SIGNAL_TYPES
]

X_train = load_X(X_train_signals_paths)
X_test = load_X(X_test_signals_paths)


# Load "y" (the neural network's training and testing outputs)

def load_y(y_path):
    file = open(y_path, 'r')
    # Read dataset from disk, dealing with text file's syntax
    y_ = np.array(
        [elem for elem in [
            row.replace('  ', ' ').strip().split(' ') for row in file
        ]], 
        dtype=np.int64
    )
    file.close()
    
    # Substract 1 to each output class for friendly 0-based indexing 
    return y_ - 1

y_train_path = os.path.join(DATASET_PATH, TRAIN, "y_train.txt")
y_test_path = os.path.join(DATASET_PATH, TEST, "y_test.txt")

y_train = load_y(y_train_path)
y_test = load_y(y_test_path)
```

```python
import torch
from torch import nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from tqdm.notebook import tqdm
from IPython.display import clear_output
```

```python
device = torch.device("cuda:5" if torch.cuda.is_available() else 'cpu')
print('Device: ', device)

class BasicDataset(Dataset):
    
    def __init__ (self, X, y):
        super(BasicDataset, self).__init__()
        
        self.X = torch.from_numpy(X)
        self.y = F.one_hot(torch.from_numpy(y).view(-1)).float()
        
    def __len__(self):
        return len(self.y)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

class SimpleNet(nn.Module):
    
    def __init__(self, in_channels=128, hidden=128, n_classes=6, num_layers=5):
        super(SimpleNet, self).__init__()
        
        self.hidden = hidden
        self.num_layers = num_layers
        self.features = nn.Sequential(
            nn.Conv1d(in_channels, 128*2, 1),
            nn.BatchNorm1d(128*2),
            nn.PReLU(128*2),
            
            nn.Conv1d(128*2, 128*3, 1),
            nn.BatchNorm1d(128*3),
            nn.PReLU(128*3),
            
            nn.Conv1d(128*3, 128*4, 1),
            nn.BatchNorm1d(128*4),
            nn.PReLU(128*4))
            
        self.lstm1 = nn.LSTMCell(input_size=9, hidden_size=hidden)
        self.lin = nn.Linear(hidden, hidden)
        self.lstm2 = nn.LSTMCell(input_size=hidden, hidden_size=hidden)
        
        self.classification = nn.Sequential(
            nn.Dropout(0.1),
            nn.Linear(hidden, 64),
            nn.PReLU(64),
            nn.Dropout(0.1),
            nn.Linear(64, 32),
            nn.PReLU(32),
            nn.Dropout(0.1),
            nn.Linear(32, n_classes))
        
    def forward(self, x):
        
        bs = x.size(0)
        h, c = self.init_state(bs)
        
        x = self.features(x)
        
        for x_t in x.transpose(1, 0):
            h, c = self.lstm1(x_t, (h, c))
            h, c = self.lstm2(self.lin(h), (h, c))

        return self.classification(h)
    
    def init_state(self, bs):
        
        return torch.randn(bs, self.hidden, requires_grad=True, device=device), \
               torch.randn(bs, self.hidden, requires_grad=True, device=device)

class СonvNet(nn.Module):
    
    def __init__(self, in_channels=9, n_classes=6):
        super(СonvNet, self).__init__()
        self.features = nn.Sequential(
            nn.Conv1d(in_channels=9, out_channels=18, kernel_size=5, padding=2),
            nn.BatchNorm1d(18),
            nn.PReLU(18), 
            nn.MaxPool1d(kernel_size=2, stride=2),
            nn.Dropout(0.2),
            
            nn.Conv1d(in_channels=18, out_channels=32, kernel_size=3),
            nn.PReLU(32),
            nn.BatchNorm1d(32),
            nn.MaxPool1d(kernel_size=2, stride=2),
            nn.Dropout())
        
        self.classification = nn.Sequential(
            nn.Linear(992, 200),
            nn.PReLU(200),
            nn.BatchNorm1d(200),
            nn.Dropout(0.2),
            nn.Linear(200, 128),
            nn.PReLU(128),
            nn.BatchNorm1d(128),
            nn.Dropout(0.2),
            nn.Linear(128, 64),
            nn.PReLU(64),
            nn.BatchNorm1d(64),
            nn.Linear(64, n_classes))
        
    def forward(self, x):
        
        bs = x.size(0)
        x = x.transpose(2,1)
        
        x = self.features(x)
        x = x.view(bs, -1)
        return self.classification(x)
    
def train_model(model, criterion, optimizer, scheduler, dataloaders_dict, num_epochs=25):
    
    since = time.time()
    model.to(device)
    best_acc = 0.0
    
    losses = {phase: [] for phase in ['train', 'val']}
    acc = {phase: [] for phase in ['train', 'val']}
    
    for epoch in tqdm(range(num_epochs)):
        
        for phase in ['train', 'val']:
            
            model.train() if phase == 'train' else model.eval()

            running_loss = 0.0
            running_corrects = 0
            
            for inputs, labels in dataloaders_dict[phase]:
                
                inputs = inputs.to(device)
                labels = labels.to(device)

                optimizer.zero_grad()
                with torch.set_grad_enabled(phase == 'train'):
                    
                    output = model(inputs)
                    _, pred = torch.max(output, 1)
                    loss = criterion(output, labels)

                    if phase == 'train':
                        loss.backward()
                        optimizer.step()
                        
                _, true = torch.max(labels, 1)
                running_corrects += torch.sum(true == pred)
                running_loss += loss.item() * inputs.size(0)
                
            if phase == 'train':
                scheduler.step()
            
            losses[phase].append(running_loss / len(dataloaders_dict[phase].dataset))
            acc[phase].append(running_corrects.double() / len(dataloaders_dict[phase].dataset))

            if phase == 'val' and acc[phase][epoch] > best_acc:
                best_acc = acc[phase][epoch]
            
        clear_output(True)
        
        for metric, name in zip((losses, acc), (' loss', ' acc')):
            
            for phase in ['train', 'val']:
                plt.plot(metric[phase], label=phase + name)

            plt.legend()
            plt.grid()
            plt.show()
            

    time_elapsed = time.time() - since
    print('Training complete in {:.0f}m {:.0f}s'.format(
        time_elapsed // 60, time_elapsed % 60))
    print('Best val Acc: {:4f}'.format(best_acc))

    return model
```

```python
bs = 10
datasets = {dataset_type: BasicDataset(X_, y_)
            for dataset_type, (X_, y_) in zip(['train', 'val'], 
                                            [(X_train, y_train), (X_test, y_test)])}

dataloaders = {dataloader_type: DataLoader(datasets[dataloader_type], batch_size=bs, shuffle=True, drop_last=True)
               for dataloader_type in ['train', 'val']}

model = SimpleNet()
optimizer = torch.optim.Adam(model.parameters(), lr=3e-3, weight_decay=1e-3)
criterion = nn.BCEWithLogitsLoss()
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=8, gamma=0.5)
```

```python
model = train_model(model, criterion, optimizer, scheduler, datasets, num_epochs=20)
```

```python
from torchvision import datasets, models, transforms
```

```python
def set_parameter_requires_grad(model, feature_extracting):
    if feature_extracting:
        for param in model.parameters():
            param.requires_grad = False
            
def initialize_model(model_name, num_classes, feature_extract=True, use_pretrained=True):
    
    def clf_block(in_size, out_size):
        return nn.Sequential(
            nn.Linear(in_size, out_size),
            nn.PReLU(out_size),
            nn.BatchNorm1d(out_size),
            nn.Dropout(0.2))

    model, coff = model_name.split('-')
    print(model, coff)
    
    if model == 'ResNet' or model == 'Wide_ResNet':
        
        if model == 'ResNet':
        
            """ 
            ResNet
            """

            if coff == '18':
                model_ft = models.resnet18(pretrained=use_pretrained)
            elif coff == '34':
                model_ft = models.resnet34(pretrained=use_pretrained)
            elif coff == '50':
                model_ft = models.resnet50(pretrained=use_pretrained)
            elif coff == '101':
                model_ft = models.resnet101(pretrained=use_pretrained)
            else:
                model_ft = models.resnet152(pretrained=use_pretrained)
        else:
            
            """
            Wide_ResNet
            """
            
            if coff == '50':
                model_ft = models.wide_resnet50_2(pretrained=use_pretrained)
            else:
                model_ft = models.wide_resnet101_2(pretrained=use_pretrained)
            
        set_parameter_requires_grad(model_ft, feature_extract)
        num_ftrs = model_ft.fc.in_features
        
        model_ft.fc = nn.Sequential(clf_block(num_ftrs, num_ftrs // 2),
                                    clf_block(num_ftrs // 2, num_ftrs // 4),
                                    clf_block(num_ftrs // 4, num_ftrs // 6),
                                    nn.Linear(num_ftrs // 6, num_classes))
                                       
        input_size = 224
        


    return model_ft.to(device), input_size

def init_dataloaders(data_dir: str, input_size: int, batch_size: int, p: float) -> dict:
    
    def transform (x):
        return F.one_hot(torch.tensor(x),  num_classes=50).float()
    
    data_transforms = {
        'train': transforms.Compose([
            transforms.RandomResizedCrop(input_size),
            transforms.RandomHorizontalFlip(),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ]),

        'val': transforms.Compose([
            transforms.Resize(input_size, interpolation=4),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ]),
    }
    
    _ = datasets.ImageFolder(os.path.join(data_dir, 'train'), data_transforms['train'], 
                             target_transform=transform)
    train, val = torch.utils.data.random_split(_, (int(len(_)*p), len(_) - int(len(_)*p)))
    
    dataloaders_dict = {x: DataLoader(dataset, batch_size=batch_size, shuffle=True, drop_last=True) 
                        for x, dataset in zip(['train', 'val'], [train, val])}
    
    return dataloaders_dict

def init_optimizer(model, lr=5e-4, w_d=5e-4, feature_extract=True):
    
    params_to_update = model.parameters()
    
    if feature_extract:
        params_to_update = []
        for name, param in model.named_parameters():
            if param.requires_grad == True:
                params_to_update.append(param)


    return torch.optim.Adam(params_to_update, lr=lr, weight_decay=w_d)
```

```python
model, input_size = initialize_model('Wide_ResNet-101', 50, use_pretrained=True)
dataloaders_dict = init_dataloaders('data', input_size, 10, 0.9)
optim = init_optimizer(model)
criterion = nn.BCEWithLogitsLoss()
scheduler = torch.optim.lr_scheduler.StepLR(optim, step_size=7, gamma=0.5)
```

```python
model = train_model(model, criterion, optim, scheduler, dataloaders_dict, num_epochs=40)
```

```python
model_res, input_size = initialize_model('ResNet-152', 50, use_pretrained=True)
dataloaders_dict = init_dataloaders('data', input_size, 10, 0.9)
optim = init_optimizer(model_res)
criterion = nn.BCEWithLogitsLoss()
scheduler = torch.optim.lr_scheduler.StepLR(optim, step_size=10, gamma=0.5)

model_res = train_model(model_res, criterion, optim, scheduler, dataloaders_dict, num_epochs=60)
```
